## Limit calculations using deterministic ODEs


```python
# steady state for mRNA level
def m(omega, k0, k1):
    return omega * (k0 / k1)
```

```python
k0 = 0.2
k1 = 0.01
omega = 1

print("mean: ", m(omega, k0, k1))
```

## Mean Expression Level from stochastic simulation


```python
from ssa import SSA
```

```python
dt = 0.5  # time-step
x = 0  # initial state
t = 0.0

ssa = SSA(k0_production_rate_const=k0, k1_degradation_rate_const=k1)

mrna = [x]
time = [t]
for i in range(10000):
    x = ssa.compute_step(x, t, t + dt)
    t += dt
    mrna.append(x)
    time.append(t)
```

```python
import matplotlib.pyplot as plt

plt.plot(time[100:], mrna[100:])
plt.show()
```

```python
import numpy as np

mean = np.mean(mrna[100:])
print("mean: ", mean)
```
